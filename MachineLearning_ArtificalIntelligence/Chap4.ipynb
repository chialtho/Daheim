First Steps in the sklearn library

```python
from sklearn import datasets
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=2)
dtc.fit(x, y)
y_pred = dtc.predict(x)
dtc.score(x, y)
```

Plot results

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.title("Plot of labels and predictions using Decision Tree Classifier")
ax1.plot(y, color = "#020640", marker = ".", label = "Labels")
ax1.plot(y_pred, color = "#05f9d4", label = "Predictions", alpha = .5)
plt.legend(loc = "upper left")
plt.grid(True)
plt.show()
```
