# Chapter 9: Probabilistic Models

## 9.1 Introduction

Two types of probabilistic models:
- Generative
  - more holistic approach
  - model joint probability $P (X; Y)$
  - model changes in the output based on changes in input **and** state
  - harder to build
  - more information and assumptions to start with
- Discriminative
  - model conditional probability $P (Y|X)$
  - predict the changes in the output based on only changes in the input 
  - do not explicitly try to understanf changes in the state of the system

where
- $X$ is the observable input 
- $Y$ the observable output
- $S$ is the state

Two different groups:

- Maximum likelihood estimation
- Bayesian approach


## 9.2 Discriminative Models

### 9.2.1 Maximum Likelihood Estimation

\begin{equation}s
\theta^{MLE} = {arg\,max}_{\theta \in \Theta} {L(y|\theta)}
\end{equation}

### 9.2.2 Bayesian Approach

\begin{equation}
P(\theta|y) = \frac{P(y|\theta) * P(\theta)}{P(y)}
\end{equation}

\begin{equation}
\theta^{Bayes} = {arg\,max}_{\theta \in \Theta} {P(\theta|y)}
\end{equation}

## 9.3 Implementing Probabilistic Models

Naive Bayes approach: different dimensions are independent of each other


```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn import naive_bayes
from sklearn.model_selection import train_test_split

nbc = naive_bayes.GaussianNB()
nbc.fit(X=X, y=y)
y_pred = nbc.predict(X=X)
nbc.score(X=X, y=y)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(y, color = "#020640", marker = ".", label = "Labels")
ax1.plot(y_pred, color = "#05f9d4", label = "Predictions", alpha = .5)
plt.title("Labels and predictions using Naive Bayes")
plt.legend(loc = "upper left")
plt.grid(True) 
plt.show()
```
